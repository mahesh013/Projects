```python
plt.figure(figsize=(10,7))

top_clubs = fifa.groupby(['club']).overall.mean().sort_values(ascending  = False)[:20]
plt.title("Top 20 Clubs in FIFA20")
sns.barplot(y = top_clubs.index, x = top_clubs, palette="Set2")
for index, value in enumerate(top_clubs):
    plt.text(value//2, index, str(round(value,2)))
    
```

```python
plt.figure(figsize=(10,7))

top_countries = fifa.groupby(['nationality']).overall.mean().sort_values(ascending  = False)[:20]
plt.title("Top 20 Countries in FIFA20")
sns.barplot(y = top_countries.index, x = top_countries, palette="cubehelix")
for index, value in enumerate(top_countries):
    plt.text(value//2, index, str(round(value,2)))
    
plt.tight_layout(pad=4.0)
```

```python
mycolor=["b","r"]
plt.pie(x = fifa.groupby(['preferred_foot']).preferred_foot.count().to_list(),
        labels = ["Left", "Right"], autopct='%1f%%', explode = (0, 0.1) , shadow=True, colors=mycolor)
```

```python
sns.histplot(x= fifa.age, kde=True, bins=20,color='black')
plt.axvline(x= fifa.age.mean(),c='red',ls='-.',label='Mean Age')
plt.legend()
plt.title('Distribution of Age')
```
